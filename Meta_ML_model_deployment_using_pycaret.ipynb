```python
import pandas as pd
from pycaret.datasets import get_data
from pycaret.classification import *
```

```python
# Load dataset
data = get_data('iris')

# Display first few rows
data.head()
```

```python
# Initialize setup
clf1 = setup(data, target='species', session_id=123, log_experiment=False, experiment_name='iris_exp', log_plots=False, log_profile=False, log_data=False)

```

```python
# Compare models
best_model = compare_models()
```

```python
# Create and tune the best model
tuned_model = tune_model(best_model)
```

```python
# Evaluate the model
evaluate_model(tuned_model)
```

```python
# Finalize the model
final_model = finalize_model(tuned_model)
```

```python
# Save the model
save_model(final_model, 'final_iris_model')
```

```python
import streamlit as st
import pandas as pd
from pycaret.classification import load_model, predict_model

# Load the model
model = load_model('final_iris_model')

# Function to predict species
def predict_species(input_data):
    data = pd.DataFrame([input_data])
    predictions = predict_model(model, data=data)
    return predictions['Label'][0]

# Streamlit app layout
st.title('Iris Species Prediction App')

sepal_length = st.number_input('Sepal Length (cm)')
sepal_width = st.number_input('Sepal Width (cm)')
petal_length = st.number_input('Petal Length (cm)')
petal_width = st.number_input('Petal Width (cm)')

input_data = {
    'sepal_length': sepal_length,
    'sepal_width': sepal_width,
    'petal_length': petal_length,
    'petal_width': petal_width
}

if st.button('Predict'):
    species = predict_species(input_data)
    st.write(f'The predicted species is: {species}')
```
